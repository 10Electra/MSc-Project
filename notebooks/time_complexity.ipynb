```python
names = (
    # ('mustard-bottle', 'mustard-bottle.obj'),
    ('table', 'table.obj'),
    # ('power-drill', 'power-drill.obj'),
    # ('bleach', 'bleach.obj'),
    # ('pitcher', 'pitcher.obj'),
    # ('mug', 'mug.obj'),
    # ('extra-large-clamp', 'extra-large-clamp-leaning.obj'),
)

gt_meshes = dict()
for foldername,filename in names:
    print(f'Getting the {foldername}')
    
    gt_mesh = o3d.io.read_triangle_mesh(f"../data/posed-meshes/{foldername}/{filename}", enable_post_processing=True)

    bake_uv_to_vertex_colours(gt_mesh)
    gt_mesh.compute_vertex_normals()

    gt_meshes[foldername] = gt_mesh

gt_mesh_list = list(gt_meshes.values())
```

```python
centres = []
for gt_meshname, gt_mesh in gt_meshes.items():
    if gt_meshname == 'table' and len(names) > 1:
        continue
    centres.append(gt_mesh.get_center())

centres = np.vstack(centres)

obj_centre = centres.mean(axis=0)
```

```python
cc1=np.array([0, 0, 1.5])

look_at = obj_centre
width: int = 370
height: int = 250
fov: float = 70.0
k: float = 5
max_normal_angle_deg = None
N = 2
include_depth_images = True

print(f'object_centre={obj_centre}')
print(f'object_centre={cc1 + np.array([0, 0, 0.5-cc1[2]])}')

def get_scans(N, width, height, ccs):
    if isinstance(ccs, np.ndarray):
        ccs_l = [ccs for _ in range(N)]
    elif isinstance(ccs, list):
        ccs_l = ccs
    else:
        raise ValueError('ccs must be array or list')
    
    pairs = []
    for cc in ccs_l:
        object_meshes, object_weights = virtual_mesh_scan(
            meshlist=gt_mesh_list,
            cam_centre=cc,
            look_at=cc+np.array([0, 0.001, -0.5]),
            k=k,
            max_normal_angle_deg=max_normal_angle_deg,
            width_px=width,
            height_px=height,
            fov=70,
            constant_uncertainty=2e-4,
            linear_uncertainty=3e-3,     # rate of uncertainty increase with depth
            quadrt_uncertainty=1e-3,     # quadratic uncertainty coefficient
            constant_perlin_sigma=2e-4,     # constant perlin noise term
            linear_perlin_sigma=3e-3,     # linear depth term
            quadrt_perlin_sigma=1e-3,     # quadratic depth term
            perlin_octaves=3,
            seg_scale_std=1e-4,      # std of per-segment scale noise
            rot_std=1e-4,      # std of global rotation noise
            trn_std=1e-3,      # std of global translation noise
            grazing_lambda=1.0,      # sigma multiplier at grazing angles; 0 disables
            seed = None,
            include_depth_image=False,
        )
        assert len(object_meshes) == 1 and len(object_weights) == 1
        pairs.append((object_meshes[0], object_weights[0]))
    return pairs

# dcols = distinct_colours(len(meshes))
# cmeshes = [copy.deepcopy(mesh) for mesh in meshes]
# for i,mesh in enumerate(cmeshes):
#     mesh.paint_uniform_color(dcols[i])
# o3d.visualization.draw_geometries(cmeshes)
```

```python
from time import perf_counter
import gc

K = 4
widths  = np.linspace(50, 400, K, dtype=int)
heights = np.linspace(50, 400, K, dtype=int)

runs_per_size = 1

results = []

for i, (width, height) in enumerate(zip(widths, heights)):
    _warm = get_scans(2, width, height, np.array([0,0,1.5]))
    m0, w0 = _warm[0]
    m1, w1 = _warm[1]
    _ = fuse_meshes(mesh1=m0, weights1=w0, mesh2=m1, weights2=w1,
                    h_alpha=3.0, r_alpha=2.5,
                    nrm_shift_iters=2, nrm_smth_iters=1,
                    sigma_theta=0.1, normal_diff_thresh=20,
                    ball_radius_percentiles=[0,1,5,10,15,25,35,50,65,75,85,90,95,99],
                    bilateral_weight_update=False, shift_all=False)

    per_call_times = []

    for r in range(runs_per_size):
        gc.collect()

        scan_mw_pairs = get_scans(2, width, height, np.array([0, 0, 1.5]))
        spf_unc_mesh, spf_unc_weights = scan_mw_pairs[0]

        for (mesh, weights) in scan_mw_pairs[1:]:
            t0 = perf_counter()
            spf_unc_mesh, spf_unc_weights = fuse_meshes(
                mesh1=spf_unc_mesh, weights1=spf_unc_weights,
                mesh2=mesh,         weights2=weights,
                h_alpha=3.0, r_alpha=2.5,
                nrm_shift_iters=2, nrm_smth_iters=1,
                sigma_theta=0.1,
                normal_diff_thresh=20,
                ball_radius_percentiles=[0,1,5,10,15,25,35,50,65,75,85,90,95,99],
                bilateral_weight_update=False,
                shift_all=False,
            )
            per_call_times.append(perf_counter() - t0)

    # summarise for this size
    per_call_times = np.array(per_call_times)
    results.append({
        "width":  int(width),
        "height": int(height),
        "N":      int(width*height),
        "calls":  len(per_call_times),
        "mean_s": per_call_times.mean(),
        "median_s": np.median(per_call_times),
        "p95_s":  np.quantile(per_call_times, 0.95),
    })

N = np.array([r["N"] for r in results], dtype=float)
T = np.array([r["mean_s"] for r in results], dtype=float)

plt.scatter(N,T)

# complexity fit: T =~ k * N^p
# mask = (N > 0) & (T > 0)
# p, logk = np.polyfit(np.log(N[mask]), np.log(T[mask]), 1)
# k = np.exp(logk)

# print(f"Empirical fit: time =~ {k:.3e} * N^{p:.2f}")
# for r in results:
#     print(f'{r["width"]}*{r["height"]}: mean {r["mean_s"]*1e3:.1f} ms '
#           f'(calls={r["calls"]}, p95 {r["p95_s"]*1e3:.1f} ms)')
```

```python
def rgb_error_maps(depth_data_gt, depth_data_recon):
    out = []
    for gt, re in zip(depth_data_gt, depth_data_recon):
        rgb_gt = np.asarray(gt['rgb'])[..., :3]
        rgb_re = np.asarray(re['rgb'])[..., :3]
        rgb_gt = np.flip(rgb_gt)
        rgb_re = np.flip(rgb_re)

        out.append(np.linalg.norm(rgb_re - rgb_gt, axis=2))
    return out
```

```python
import os

def plot_images(images, max_cols=5, titles=None, per_img_width=3, cmap=None, cbar_cols=None, cbar_kw=None):
    from mpl_toolkits.axes_grid1 import make_axes_locatable
    n = len(images)
    h_to_w = images[0].shape[0] / images[0].shape[1]
    cols = max(1, min(max_cols, n))
    rows = (n + cols - 1) // cols
    fig, axes = plt.subplots(rows, cols, figsize=(cols*per_img_width, rows*per_img_width*h_to_w), squeeze=False)
    axes = axes.ravel()
    for i, ax in enumerate(axes):
        if i >= n: ax.axis('off'); continue
        img = images[i]
        arr = plt.imread(img) if isinstance(img, (str, os.PathLike)) else np.asarray(img)
        im = ax.imshow(arr, cmap=(cmap if arr.ndim == 2 else None))
        ax.set_xticks([]); ax.set_yticks([])
        for s in ax.spines.values(): s.set_visible(False)
        if titles and i < len(titles) and titles[i]: ax.set_title(titles[i], fontsize=9, pad=2)
        if arr.ndim == 2 and cbar_cols and (i % cols) in cbar_cols:
            cax = make_axes_locatable(ax).append_axes("right", size="5%", pad=0.02)
            fig.colorbar(im, cax=cax, **(cbar_kw or {}))
    plt.subplots_adjust(left=0, right=1, top=1, bottom=0, wspace=0, hspace=0)
    return fig, axes[:n]
```

```python
plot_images(rgb_error_maps_tsdf, max_cols=4)
plot_images(rgb_error_maps_unc, max_cols=4)
```
